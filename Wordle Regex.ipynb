```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
```

```python
valid_solutions_df = pd.read_csv('solution_set.csv')
valid_solutions_df
```

```python
# first letter is S, followed by four letters in EAROTL in any order
pattern = r's[earotl]{4}'
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```

```python
# The letters in ALINC, in any order
pattern = r'[alinc]{5}'
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```

```python
# None of the letters in STORECANL
# followed by A then N
# followed by exactly two letters that are not in STORENL
pattern = r'[^storecanl]an[^storeanl]{2}' 
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```

```python
# [^storecanl] - First letter is not in STORECANL
# an - Second and third letters are AN
# ([^storeanl]c|c[^storeanl]) - last two letters are either
#  * any letter not in STORECANL followed by a C, OR
#  * a C followed by any letter not in STORECANL
pattern = r'[^storecanl]an([^storeanl]c|c[^storeanl])'
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```

### AND LOGIC

```python
# s - first letter is definitely S
# [^te] - second letter is not T or E
# [^teo] - third letter is not T, E, or O
# [^ter] - third letter is not T, E, or R
# [^te] - fifth letter is not T or E
pattern = r's[^te][^teo][^ter][^te]'
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```

```python
# s - first letter is definitely S
# [^te] - second letter is not T or E
# [^teo] - third letter is not T, E, or O
# [^ter] - third letter is not T, E, or R
# [^te] - fifth letter is not T or E
# (?=.*o.*) - O must appear somewhere
# (?=.*r.*) - R must appear somewhere
pattern = r's(?=[^te][^teo][^ter][^te])(?=.*o.*)(?=.*r.*)'
matching_words = valid_solutions_df.loc[valid_solutions_df['word'].str.match(pattern)]
matching_words
```
